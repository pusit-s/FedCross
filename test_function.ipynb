```python
import os
from pathlib import Path
import shutil

def move_and_rename_segmentation_files(base_path, label_folder_name='label'):
    # Ensure base_path is a Path object
    base_path = Path(base_path)
    
    # Create a Path object for the label folder
    label_path = base_path / label_folder_name
    
    # Create the 'label' directory if it does not exist
    label_path.mkdir(exist_ok=True)
    
    # Pattern to match segmentation files
    pattern = "*_segmentation*"
    
    # Move all segmentation files to the 'label' directory and rename them
    for seg_file in base_path.glob(pattern):
        # Define the new name by removing '_segmentation'
        new_name = seg_file.name.replace('_segmentation', '')
        # Define the new path for the segmentation file within the label directory
        new_path = label_path / new_name
        
        # Move and rename the file
        shutil.move(str(seg_file), new_path)
        
        # Optionally, you can print out the name of the files that have been moved and renamed
        print(f"Moved and renamed segmentation file: {seg_file.name} to {new_path}")

# Use the function for the desired directory
move_and_rename_segmentation_files('dataset/PROMISE12/training_data')
```

```python
import os
from pathlib import Path
import shutil

def move_segmentation_files(base_path, label_folder_name='label'):
    # Ensure base_path is a Path object
    base_path = Path(base_path)
    
    # Create a Path object for the label folder
    label_path = base_path / label_folder_name
    
    # Create the 'label' directory if it does not exist
    label_path.mkdir(exist_ok=True)
    
    # Pattern to match segmentation files
    pattern = "*_segmentation*"
    
    # Move all segmentation files to the 'label' directory
    for seg_file in base_path.glob(pattern):
        # Define the new path for the segmentation file
        new_path = label_path / seg_file.name
        
        # Move the file
        shutil.move(str(seg_file), new_path)
        
        # Optionally, you can print out the name of the files that have been moved
        print(f"Moved segmentation file: {seg_file.name} to {label_folder_name}/")

# Use the function for the desired directory
move_segmentation_files('dataset/PROMISE12/image')
```

```python
import os
from pathlib import Path

def rename_segmentation_files(folder_path):
    # Ensure folder_path is a Path object
    folder_path = Path(folder_path)
    
    # Check if the folder exists
    if not folder_path.exists() or not folder_path.is_dir():
        print(f"The directory {folder_path} does not exist.")
        return
    
    # Pattern to match segmentation files
    pattern = "*_segmentation*"
    
    # Rename files in the specified folder
    for file_path in folder_path.glob(pattern):
        # Define the new name by removing '_segmentation'
        new_name = file_path.name.replace('_segmentation', '')
        # Define the new path for the file within the same directory
        new_path = file_path.with_name(new_name)
        
        # Rename the file
        file_path.rename(new_path)
        
        # Optionally, you can print out the name of the files that have been renamed
        print(f"Renamed file: {file_path.name} to {new_name}")

# Use the function for the desired folder
rename_segmentation_files('dataset/PROMISE12/label')
```
